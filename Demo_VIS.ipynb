```python
from modules.visionllms import QwenVL
from utils import calculate_accuracy_metrics
import json
```

```python
system_prompt = """
You are an expert in the field of object detection in images according to the request,
process the input image and return a list of bounding boxes in JSON format.
For example:
{
"bbox":[[x1_1, y1_1, x2_1, y2_1], [x1_2, y1_2, x2_2, y2_2]]
}
"""
qwen = QwenVL(system_prompt=system_prompt)
```

```python
detect_boxes = {'cables': [[49, 216, 140, 318]],
 'keyboard': [[165, 273, 338, 322]],
 'table': [[130, 231, 639, 477]],
 'picture': [[428, 110, 514, 194], [210, 0, 271, 42]],
 'girl': [[525, 95, 638, 233]],
 'monitor': [[157, 126, 305, 271]],
 'hair': [[582, 98, 641, 210]],
 'chain': [[593, 170, 615, 178]],
 'desk': [[532, 160, 599, 216]],
 'computer': [[413, 66, 494, 113], [216, 351, 295, 480]],
 'wall': [[252, 62, 493, 113], [0, 0, 540, 86]],
 'computer mouse': [[345, 300, 374, 323]],
 'phone': [[456, 271, 551, 348]],
 'bag': [[574, 275, 640, 405]],
 'pen': [[405, 324, 458, 334]],
 'cable': [[141, 372, 223, 466]],
 'drawer': [[3, 424, 133, 467]],
 'floor': [[110, 339, 616, 480]],
 'outlet': [[86, 202, 98, 223]]}
```

```python
desc_queries = [
'cables to the left of keyboard',
'cables to the left of table',
'picture to the left of girl',
'picture to the right of monitor',
'picture to the left of hair',
'picture to the left of chain',
'picture to the left of desk',
'computer on wall',
'computer behind girl',
'hair to the right of desk',
'hair to the right of picture',
'computer mouse to the left of phone',
'computer mouse next to keyboard',
'computer mouse near keyboard',
'computer mouse to the right of keyboard',
'bag to the right of phone',
'bag to the right of pen',
'cable to the left of computer',
'cable to the right of drawer',
'cable on floor',
'floor to the right of drawer',
'pen to the left of phone',
'pen to the left of bag',
'pen next to phone',
'pen near phone',
'drawer to the left of floor',
'drawer to the left of cable',
'drawer to the left of computer',
'drawer to the left of table',
'outlet to the left of monitor',
'outlet on wall',
'picture on wall',
'monitor behind keyboard',
'monitor to the right of outlet',
'monitor to the left of picture',
'computer to the right of cable',
'computer to the right of drawer',
'desk to the left of chain',
'desk to the left of hair',
'desk to the right of picture',
'phone near pen',
'phone to the right of keyboard',
'phone on table',
'phone to the left of bag',
'phone to the right of computer mouse',
'phone to the right of pen',
'chain to the right of picture',
'chain to the right of desk',
'table to the right of cables',
'table to the right of drawer',
'keyboard in front of monitor',
'keyboard near computer mouse',
'keyboard to the left of phone',
'keyboard to the right of cables',
'keyboard to the left of computer mouse',
'girl sitting at desk',
'girl to the right of picture',
'girl in front of computer',
'girl working on computer',
]
```

```python
preds = []
preds_rel = {}
truths_rel = {}
```

```python
for desc_query in desc_queries:
  output_txt = qwen.process(
      img_path="3.jpg",
      user_prompt=f"""
  with those given bounding box:
  {str(detect_boxes)}
  extract the bounding boxes that match this description:
  {desc_query}
      """
      )
  bbox = json.loads(output_txt[0])
  box_key = list(bbox.keys())[0]
  preds.append({
      "bbox": bbox[box_key],
      "rel":desc_query
  })
```

```python
with open("predictions.json", "w", encoding="utf-8") as json_file:
    json.dump(preds, json_file, indent=4, ensure_ascii=False)
```

```python
for result in preds:
  if result['rel'] not in preds_rel:
    preds_rel[result['rel']] = []
  preds_rel[result['rel']].extend(result['bbox'])
```

```python
with open("img_rels.json", "r") as file:
    truths = json.load(file)
```

```python
for truth in truths:
  if truth['rel'] not in truths_rel:
    truths_rel[truth['rel']] = []
  truths_rel[truth['rel']].append(truth['bboxes'])
```

```python
# Calculate accuracy metrics with default IoU threshold of 0.5
metrics = calculate_accuracy_metrics(truths_rel, preds_rel)

# Print results
print("\nPer-class metrics:")
for class_name, class_metrics in metrics['per_class'].items():
    print(f"\n{class_name}:")
    for metric_name, value in class_metrics.items():
        print(f"  {metric_name}: {value:.3f}")
```

```python
print("\nOverall metrics:")
for metric_name, value in metrics['overall'].items():
    print(f"  {metric_name}: {value:.3f}")
```
